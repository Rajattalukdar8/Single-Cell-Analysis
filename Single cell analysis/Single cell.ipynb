# Data preprocessing 

```python
#checking highly expression genes in cancer and normal subject
sc.pl.highest_expr_genes(normal, n_top = 10)
```

```python
sc.pl.highest_expr_genes(normal, n_top = 10)
```

# checking highly expression genes in cancer and normal subject

```python
sc.pp.filter_cells(normal, min_genes = 200) #filter out cell where minimum genes epressed is less than 200
```

```python
sc.pp.filter_cells(cancer, min_genes = 200) 
```

```python
sc.pp.filter_genes(normal, min_cells = 3)#filter out genes which are expressed in less than 3 cell
```

```python
sc.pp.filter_genes(cancer, min_cells = 3) 
```

# Quality control

```python
#filtering mitochondril gene
normal.var['mt'] = normal.var_names.str.startswith('MT-')
sc.pp.calculate_qc_metrics(normal, qc_vars=['mt'],percent_top=None, inplace=True)
```

```python
cancer.var['mt'] = cancer.var_names.str.startswith('MT-')
sc.pp.calculate_qc_metrics(cancer, qc_vars=['mt'],percent_top=None, inplace=True)
```

```python
sc.pl.violin(normal,['n_genes_by_counts','total_counts', 'pct_counts_mt'],  
             jitter=0.4, multi_panel=True)
```

```python
sc.pl.violin(cancer,['n_genes_by_counts','total_counts', 'pct_counts_mt'],
             jitter=0.4, multi_panel=True)
```

```python
#removing the genes which are highly expressed
normal=  normal[normal.obs.n_genes_by_counts < 2500,:]
cancer = cancer[cancer.obs.n_genes_by_counts < 2500,:]
```

```python
#filter cell having greater than 10% mitochondrial count
normal = normal[normal.obs.pct_counts_mt <10,:]
```

```python
cancer =cancer[cancer.obs.pct_counts_mt <10,:] 
```

# log normalization

```python
sc.pp.normalize_total(normal, target_sum=1e4)
sc.pp.log1p(normal) 
```

```python
sc.pp.normalize_total(cancer, target_sum=1e4)
sc.pp.log1p(cancer)
```

# selecting highly variable genes

```python
sc.pp.highly_variable_genes(normal, n_top_genes = 1000)
normal = normal[:,normal.var.highly_variable]
```

```python
sc.pp.highly_variable_genes(cancer, n_top_genes = 1000)
cancer = cancer[:,cancer.var.highly_variable]
```

# Scaling

```python
sc.pp.scale(normal, max_value=10)  
sc.pp.scale(cancer, max_value=10)  
```

```python
normal.to_df()
```

```python
cancer.to_df()
```
